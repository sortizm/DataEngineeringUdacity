## World happiness
This data is from the report World Happiness report. The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale.

Origin: https://www.kaggle.com/unsdsn/world-happiness

```python
df = pd.read_csv("../data/world_happiness_2016.csv")
print(df.shape[0])
df.info()
```
