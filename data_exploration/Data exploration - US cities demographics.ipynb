# Data exploration - US cities demographics

```python
import pandas as pd
```

## US cities demographics

This dataset contains information about the demographics of all US cities and census-designated places with a population greater or equal to 65,000. 


Origin: https://public.opendatasoft.com/explore/dataset/us-cities-demographics/export/

## Data info

```python
df = pd.read_csv("../data/us-cities-demographics.csv", sep=';')
print(f"Rows: {df.shape[0]}")

df.info()
df.sort_values('City').head(10)
```

```python
df['race_population_percentage'] = df['Count'] / df['Total Population']
```

```python
df[['State', 'City']].groupby('State', as_index=False).count().sort_values('City', ascending=False).head(10)
```

```python
df.groupby('State', as_index=False)['Count'].sum().sort_values('Count', ascending=False).head(10)
```

```python
hispanic_df = df[df.Race == 'Hispanic or Latino']
hispanic_df.groupby('State', as_index=False)['Count'].sum().sort_values('Count', ascending=False).head(10)
```

```python
hispanic_df.groupby('State', as_index=False)['race_population_percentage'].sum().sort_values('race_population_percentage', ascending=False).head(10)
```

```python
df.groupby(['City', 'State'], as_index=False)['Total Population'].sum().sort_values('Total Population', ascending=False).head(10)
```
